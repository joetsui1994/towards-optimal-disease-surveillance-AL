#### load in modules

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import to_rgba
```

#### colour scheme

```python
# define the colors for the different policies
policy_colors = {
    'random': '#151515',
    'reactive_infected': '#6494AA',
    'degree_central': '#C47B93',
    'pagerank_central': '#7072B1',
    'node_entropy': '#BE953E',
    'bald': '#90A959',
    'le': '#9A3526'
}
```

#### test budget availability: low

```python
# specify data
top_ranking_policies = np.array(
    [['le', 'le', 'le'],
     ['pagerank_central', 'le', 'le'],
     ['pagerank_central', 'le', 'le'],
     ['degree_central', 'degree_central', 'pagerank_central']]
     )
# these values are generated using the AUC_plots.ipynb notebook for each outbreak scenario
top_ranking_freqs = np.array(
    [[0.3593230769230769, 0.43777777777777777, 0.4548878048780488],
     [0.33128571428571435, 0.372992, 0.409545945945946],
     [0.3041066666666667, 0.30806451612903224, 0.32504186046511624],
     [0.348295652173913, 0.36824761904761905, 0.241112]]
     )
```

```python
# summary of top-ranking policies and their top-ranking frequencies on under different outbreak scenarios and with different test budgets
fig, ax = plt.subplots(figsize=(4.5, 7))

# labels for different synthetic networks
graph_labels = ['Periodic Lattice', 'Stochastic Block\n(high-modularity)', 'Stochastic Block\n(low-modularity)', 'Barabasi-Albert']
# labels for different stages of outbreak progression
outbreak_stages = ['Early\n(I/N=0.1)', 'Middle\n(I/N=0.3)', 'Late\n(I/N=0.5)']

# map policy names to colors
policy_colours = np.array([[policy_colors[policy] for policy in row] for row in top_ranking_policies])
policy_colours_rgb = np.array([[to_rgba(policy_colours[i][j], alpha=0.4) for j in range(len(top_ranking_policies[i]))]
                               for i in range(len(top_ranking_policies))])

# colour cells based on corresponding top-ranking policy
im = ax.imshow(policy_colours_rgb)

# show all ticks and label them with the respective list entries
ax.set_xticks(np.arange(len(outbreak_stages)), labels=outbreak_stages)
ax.set_yticks(np.arange(len(graph_labels)), labels=graph_labels)

# rotate the tick labels and set their alignment, add vertical padding
plt.setp(ax.get_xticklabels(), rotation=90, ha="right", rotation_mode="anchor", va="center_baseline")

# add circle to each cell of certain radius
for i in range(len(graph_labels)):
    for j in range(len(outbreak_stages)):
        circle = plt.Circle((j, i), 0.45 * top_ranking_freqs[i][j], facecolor=policy_colours[i][j], fill=True, lw=1.5)
        ax.add_artist(circle)

# add thick border
for _, spine in ax.spines.items():
    spine.set_visible(True)
    spine.set_linewidth(2)

# make ticks thicker
ax.tick_params(axis='both', which='major', width=2)

# adjust axis tick label size
ax.tick_params(axis='both', which='major', labelsize=13)

# use tight layout
fig.tight_layout()

plt.show()
```
