#### load in modules

```python
import networkx as nx
import matplotlib.pyplot as plt
```

#### generate random graph using the Barabasi-Albert model

```python
# params
num_nodes = 160
min_degree = 2

# generate graph
ba_G = nx.barabasi_albert_graph(num_nodes, min_degree)
```

```python
# draw graph
plt.figure(figsize=(5, 5))
pos = nx.spring_layout(ba_G)
nx.draw(ba_G, pos, node_size=50, node_color='k', with_labels=False)
plt.show()
```

```python
# export graph as adjlist
nx.write_adjlist(ba_G, 'BA-m2.adjlist')
```
