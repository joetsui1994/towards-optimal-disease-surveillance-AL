#### load in modules

```python
import networkx as nx
import matplotlib.pyplot as plt
```

#### create aperiodic lattice graphs

```python
# use grid_2d_graph
dim = 13
aperiodic_lattice_G = nx.grid_2d_graph(dim, dim)
```

```python
# draw graph
plt.figure(figsize=(5, 5))

# relabel nodes to 1D
mapping = { (x,y): x*dim+y for x, y in aperiodic_lattice_G.nodes() }
aperiodic_lattice_G = nx.relabel_nodes(aperiodic_lattice_G, mapping)
pos = { node: (node % dim, node // dim) for node in aperiodic_lattice_G.nodes }
nx.draw(aperiodic_lattice_G, pos, node_color='k', node_size=20, width=0.5)
plt.show()
```

```python
# export graph as adjlist
nx.write_adjlist(aperiodic_lattice_G, 'aperiodic_lattice.adjlist')
```

#### create periodic lattice graphs

```python
# use grid_2d_graph
dim = 13
periodic_lattice_G = nx.grid_2d_graph(dim, dim, periodic=True)
```

```python
# draw graph
plt.figure(figsize=(5, 5))
nx.draw(periodic_lattice_G, node_color='k', node_size=20, width=0.5)
plt.show()
```

```python
# export graph as adjlist
nx.write_adjlist(periodic_lattice_G, 'periodic_lattice.adjlist')
```
